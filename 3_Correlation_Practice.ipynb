#Business Understanding
This analysis aims to evaluate the relationship between RAspend and Sales to determine the effectiveness of radio marketing investments. By assessing the correlation between these variables, the study seeks to provide evidence on whether increases in advertising expenditure are associated with corresponding improvements in sales performance. The findings will support data-driven decisions on budget allocation and the strategic optimization of marketing initiatives.

```python
# Import the necessary libraries
%pip install qdesc
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import qdesc as qd
```

```python
# Loading our data set RadioAds.xlsx
df = pd.read_excel("RadioAds.xlsx")
df.head()
```

```python
# View the information of the dataframe.
df.info()
```

```python
# Generating a quick descriptive analytics of the variables.
desc = df.describe()
desc.T
```

```python
# Doing a normality check using Data Visualizations and AD-test
qd.normcheck_dashboard(df)
```

```python
# Choose the right approach for correlation of review duration and grades given the result of the EDA.
pearson_corr = df['RAspend'].corr(df['Sales'], method='pearson')
spearman_corr = df['RAspend'].corr(df['Sales'], method='spearman')

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```

```python
# Create a scatterplot to visualize the relationship between review duration and grades.
plt.figure(figsize=(8,5))
plt.scatter(df['RAspend'], df['Sales'])
plt.xlabel('RAspend')
plt.ylabel('Sales')
plt.title('Scatterplot of RAspend vs Sales')
plt.grid(True)
plt.show()
```
